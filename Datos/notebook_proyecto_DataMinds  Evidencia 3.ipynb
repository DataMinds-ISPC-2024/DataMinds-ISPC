5-Detección de datos en formato incorrecto

```python
# Verificar los tipos de datos, usando el metodo dtypes encontraremos los siguientes tipos de datos
#1)object: Valores de texto o numéricos mixtos.
#2)bool: Valores verdaderos o falsos.
#3)int64: Valores enteros.
#4)float64: Valores de punto flotante.
#5)datetime64: Valores de fecha y hora
print(df.dtypes)
```

```python
# Convertir tipos de datos después de cargar los datos, aqui arreglamos el tipo de dato segun lo que corresponda
df['Edad'] = pd.to_numeric(df['Edad'], errors='coerce')
df['Renunció'] = df['Renunció'].astype(bool)
df['Distancia a su casa (km)'] = pd.to_numeric(df['Distancia a su casa (km)'], errors='coerce')
df['Numero de empleado'] = pd.to_numeric(df['Numero de empleado'], errors='coerce').astype('Int64')
df['Satifacción laboral general'] = pd.to_numeric(df['Satifacción laboral general'], errors='coerce').astype('Int64')
df['Porcentaje Aumento Del Salario'] = pd.to_numeric(df['Porcentaje Aumento Del Salario'], errors='coerce')
df['Satisfacción de relaciones'] = pd.to_numeric(df['Satisfacción de relaciones'], errors='coerce').astype('Int64')
df['Cantidad de horas trabajadas'] = pd.to_numeric(df['Cantidad de horas trabajadas'], errors='coerce')
df['Antigüedad laboral'] = pd.to_numeric(df['Antigüedad laboral'], errors='coerce').astype('Int64')
df['Antigüedad en la empresa'] = pd.to_numeric(df['Antigüedad en la empresa'], errors='coerce').astype('Int64')
df['Capacitaciones del año anterior'] = pd.to_numeric(df['Capacitaciones del año anterior'], errors='coerce').astype('Int64')
df['Antigüedad en el mismo puesto'] = pd.to_numeric(df['Antigüedad en el mismo puesto'], errors='coerce').astype('Int64')
df['Años desde el ultimo ascenso'] = pd.to_numeric(df['Años desde el ultimo ascenso'], errors='coerce').astype('Int64')
df['Años con el mismo gerente'] = pd.to_numeric(df['Años con el mismo gerente'], errors='coerce').astype('Int64')

# Verificar los tipos de datos
print(df.dtypes)
```

## 2) GRÁFICOS
#a) Gráfico de torta

```python
# Filtrar el DataFrame para incluir solo 'Masculino' y 'Femenino' en la columna 'Género'
df_genero_filt = df[df['Género'].isin(['Masculino', 'Femenino'])]

# Corregir errores tipográficos
df['Género'] = df['Género'].replace({'Fem': 'Femenino', 'Femeni': 'Femenino'})

# Contar el número de ocurrencias de cada categoría
conteo_genero = df_genero_filt['Género'].value_counts()


# Para crear un gráfico de torta usamos el metodo pie
plt.figure(figsize=(8, 6)) #ajusta el tamaño del grafico, sera de 8 pulgadas por 6 pulgadas
plt.pie(conteo_genero, labels=conteo_genero.index, autopct='%1.1f%%', startangle=0) #labels: Aquí proporcionas las etiquetas para cada segmento del gráfico['Hombres', 'Mujeres'
plt.title('Distribución de género de los encuestados')                              #autopct: este parámetro controla cómo se muestran los porcentajes en cada segmento, usamos'%1.1f%%', para mostrar el porcentaje con un decimal
plt.axis('equal')  # Hace que el gráfico sea un círculo                             #startangle: Define el ángulo inicial para el primer segmento.




plt.show()
```

# b) Histograma

```python
# Crear el histograma de la columna "Edad"
plt.figure(figsize=(8, 6)) #ajusta el tamaño del grafico, sera de 8 pulgadas por 6 pulgadas
plt.hist(df['Edad'], bins=10, color='skyblue', edgecolor='black') #Los parámetros bins, color y edgecolor controlan el número de intervalos (barras), el color de las barras y el color del borde, respectivamente.
plt.title('Frecuencia de distribución de la Edad de los encuestados')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')

# Mostrar el gráfico
plt.show()
```

#c) Gráfico de dispersión

```python
# Eliminar filas con valores NaN
df.dropna(subset=['Antigüedad laboral'], inplace=True)

# Extraer las variables para el gráfico
edad = df['Edad']
antiguedad = df['Antigüedad laboral']

# Crear el gráfico de dispersión
plt.figure(figsize=(8, 6)) #ajusta el tamaño del grafico, sera de 8 pulgadas por 6 pulgadas
plt.scatter(edad, antiguedad, color='blue', s=100) #plt.scatter crear un gráfico de dispersión. El parámetro color='blue' establece el color de los puntos en el gráfico (en este caso, azul), y s=100 define el tamaño de los puntos

# Añadir título y etiquetas de los ejes
plt.title('Gráfico de Dispersión: Edad vs. Antigüedad Laboral')
plt.xlabel('Edad')
plt.ylabel('Años de Antigüedad Laboral')

# Mostrar el gráfico
plt.grid(True) #se utiliza para configurar las líneas de la cuadrícula en un gráfico. El metodo plt.grid(True) muestra la cuadricula.
plt.show()

```

## 4) Medidas descriptivas de las variables en base a su tipo (Numéricas o Categóricas). Realizar los informes según sean los datos (simétricos o asimétricos).

```python
# Medidas descriptivas para variables numéricas. El metodo describe en Pandas se utiliza para obtener estadísticas descriptivas básicas de un DataFrame
# o una serie de valores numéricos. Proporciona información como la media, la desviación estándar, los valores máximos y mínimos, y los cuartiles de los datos numéricos
print("Medidas descriptivas para variables numéricas:")
print(df.describe())
```

```python
# Medidas descriptivas para variables categóricas. El metodo describe en Pandas se utiliza para obtener estadísticas descriptivas básicas de un DataFrame
# o una serie de valores numéricos. Proporciona información como la media, la desviación estándar, los valores máximos y mínimos, y los cuartiles de los datos numéricos
columnas_categoricas = ['Viaja por trabajo', 'Departamento', 'Nivel de educacion', 'Campo de estudio', 'Entorno laboral', 'Género', 'Rol laboral', 'Estado civil']
print("\nMedidas descriptivas para variables categóricas:")
for columna in columnas_categoricas:
    print(f"\n{columna}:")
    print(df[columna].describe())
```

```python
# Informe para datos simétricos/asimétricos (usando la skewness): proporciona información sobre la forma de la distribución de los datos.
# Si es positiva, la cola derecha es más larga (distribución sesgada hacia la derecha),
# mientras que si es negativa, la cola izquierda es más larga (distribución sesgada hacia la izquierda)
print("\nInforme para datos simétricos/asimétricos:")
for columna in df.select_dtypes(include='number').columns:
    skewness = df[columna].skew()
    if skewness == 0:
        print(f"\nLos datos de '{columna}' son simétricos.")
    elif skewness > 0:
        print(f"\nLos datos de '{columna}' son asimétricos hacia la derecha (positiva).")
    else:
        print(f"\nLos datos de '{columna}' son asimétricos hacia la izquierda (negativa).")
```
