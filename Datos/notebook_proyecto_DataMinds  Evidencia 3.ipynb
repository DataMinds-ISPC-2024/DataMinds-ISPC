```python
# Convertir tipos de datos despus de cargar los datos, aqui arreglamos el tipo de dato segun lo que corresponda
df['Edad'] = pd.to_numeric(df['Edad'], errors='coerce')
df['Renunci'] = df['Renunci'].astype(bool)
df['Distancia a su casa (km)'] = pd.to_numeric(df['Distancia a su casa (km)'], errors='coerce')
df['Numero de empleado'] = pd.to_numeric(df['Numero de empleado'], errors='coerce').astype('Int64')
df['Satifaccin laboral general'] = pd.to_numeric(df['Satifaccin laboral general'], errors='coerce').astype('Int64')
df['Porcentaje Aumento Del Salario'] = pd.to_numeric(df['Porcentaje Aumento Del Salario'], errors='coerce')
df['Satisfaccin de relaciones'] = pd.to_numeric(df['Satisfaccin de relaciones'], errors='coerce').astype('Int64')
df['Cantidad de horas trabajadas'] = pd.to_numeric(df['Cantidad de horas trabajadas'], errors='coerce')
df['Antigedad laboral'] = pd.to_numeric(df['Antigedad laboral'], errors='coerce').astype('Int64')
df['Antigedad en la empresa'] = pd.to_numeric(df['Antigedad en la empresa'], errors='coerce').astype('Int64')
df['Capacitaciones del ao anterior'] = pd.to_numeric(df['Capacitaciones del ao anterior'], errors='coerce').astype('Int64')
df['Antigedad en el mismo puesto'] = pd.to_numeric(df['Antigedad en el mismo puesto'], errors='coerce').astype('Int64')
df['Aos desde el ultimo ascenso'] = pd.to_numeric(df['Aos desde el ultimo ascenso'], errors='coerce').astype('Int64')
df['Aos con el mismo gerente'] = pd.to_numeric(df['Aos con el mismo gerente'], errors='coerce').astype('Int64')

# Verificar los tipos de datos
print(df.dtypes)
```

## 2) GRFICOS
#a) Grfico de torta

```python
# Filtrar el DataFrame para incluir solo 'Masculino' y 'Femenino' en la columna 'Gnero'
df_genero_filt = df[df['Gnero'].isin(['Masculino', 'Femenino'])]

# Corregir errores tipogrficos
df['Gnero'] = df['Gnero'].replace({'Fem': 'Femenino', 'Femeni': 'Femenino'})

# Contar el nmero de ocurrencias de cada categora
conteo_genero = df_genero_filt['Gnero'].value_counts()


# Para crear un grfico de torta usamos el metodo pie
plt.figure(figsize=(8, 6)) #ajusta el tamao del grafico, sera de 8 pulgadas por 6 pulgadas
plt.pie(conteo_genero, labels=conteo_genero.index, autopct='%1.1f%%', startangle=0) #labels: Aqu proporcionas las etiquetas para cada segmento del grfico['Hombres', 'Mujeres'
plt.title('Distribucin de gnero de los encuestados')                              #autopct: este parmetro controla cmo se muestran los porcentajes en cada segmento, usamos'%1.1f%%', para mostrar el porcentaje con un decimal
plt.axis('equal')  # Hace que el grfico sea un crculo                             #startangle: Define el ngulo inicial para el primer segmento.




plt.show()
```

# b) Histograma

```python
# Crear el histograma de la columna "Edad"
plt.figure(figsize=(8, 6)) #ajusta el tamao del grafico, sera de 8 pulgadas por 6 pulgadas
plt.hist(df['Edad'], bins=10, color='skyblue', edgecolor='black') #Los parmetros bins, color y edgecolor controlan el nmero de intervalos (barras), el color de las barras y el color del borde, respectivamente.
plt.title('Frecuencia de distribucin de la Edad de los encuestados')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')

# Mostrar el grfico
plt.show()
```

#c) Grfico de dispersin

```python
# Eliminar filas con valores NaN
df.dropna(subset=['Antigedad laboral'], inplace=True)

# Extraer las variables para el grfico
edad = df['Edad']
antiguedad = df['Antigedad laboral']

# Crear el grfico de dispersin
plt.figure(figsize=(8, 6)) #ajusta el tamao del grafico, sera de 8 pulgadas por 6 pulgadas
plt.scatter(edad, antiguedad, color='blue', s=100) #plt.scatter crear un grfico de dispersin. El parmetro color='blue' establece el color de los puntos en el grfico (en este caso, azul), y s=100 define el tamao de los puntos

# Aadir ttulo y etiquetas de los ejes
plt.title('Grfico de Dispersin: Edad vs. Antigedad Laboral')
plt.xlabel('Edad')
plt.ylabel('Aos de Antigedad Laboral')

# Mostrar el grfico
plt.grid(True) #se utiliza para configurar las lneas de la cuadrcula en un grfico. El metodo plt.grid(True) muestra la cuadricula.
plt.show()

```

## 4) Medidas descriptivas de las variables en base a su tipo (Numricas o Categricas). Realizar los informes segn sean los datos (simtricos o asimtricos).

```python
# Medidas descriptivas para variables numricas. El metodo describe en Pandas se utiliza para obtener estadsticas descriptivas bsicas de un DataFrame
# o una serie de valores numricos. Proporciona informacin como la media, la desviacin estndar, los valores mximos y mnimos, y los cuartiles de los datos numricos
print("Medidas descriptivas para variables numricas:")
print(df.describe())
```

```python
# Medidas descriptivas para variables categricas. El metodo describe en Pandas se utiliza para obtener estadsticas descriptivas bsicas de un DataFrame
# o una serie de valores numricos. Proporciona informacin como la media, la desviacin estndar, los valores mximos y mnimos, y los cuartiles de los datos numricos
columnas_categoricas = ['Viaja por trabajo', 'Departamento', 'Nivel de educacion', 'Campo de estudio', 'Entorno laboral', 'Gnero', 'Rol laboral', 'Estado civil']
print("\nMedidas descriptivas para variables categricas:")
for columna in columnas_categoricas:
    print(f"\n{columna}:")
    print(df[columna].describe())
```

```python
# Informe para datos simtricos/asimtricos (usando la skewness): proporciona informacin sobre la forma de la distribucin de los datos.
# Si es positiva, la cola derecha es ms larga (distribucin sesgada hacia la derecha),
# mientras que si es negativa, la cola izquierda es ms larga (distribucin sesgada hacia la izquierda)
print("\nInforme para datos simtricos/asimtricos:")
for columna in df.select_dtypes(include='number').columns:
    skewness = df[columna].skew()
    if skewness == 0:
        print(f"\nLos datos de '{columna}' son simtricos.")
    elif skewness > 0:
        print(f"\nLos datos de '{columna}' son asimtricos hacia la derecha (positiva).")
    else:
        print(f"\nLos datos de '{columna}' son asimtricos hacia la izquierda (negativa).")
```
